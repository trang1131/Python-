## Câu lệnh if…elif…else

# Đề bài:

```python
'''
Viết chương trình nhập vào số ngày trong tuần. In ra từ tiếng anh tương ứng vs số ngày đó
Input - Output
2 - "Monday"
3 - "Tuesday"
4 - "Wednesday"
5 - "Thursday"
6 - "Friday"
7 - "Saturday"
8 - "Sunday"
'''
weekday = int(input("Please input weekday: "))
if weekday == 2:
    print("Monday")
elif weekday == 3:
    print("Tues")
elif weekday == 4:
    print("Wed")
elif weekday == 5:
    print("Thurs")
elif weekday == 6:
    print("Fri")
elif weekday == 7:
    print ("Sat")
elif weekday == 8:
    print ("Sun")
else:
    print("Số không khả dụng, vui lòng chỉ input số từ 2 tới 8")
```

### Thực hành: 
Viết chương trình nhập vào số tháng trong năm và cho biết tháng đó có bao nhiêu ngày

```python
'''
Viết chương trình nhập vào s vào số tháng trong năm và cho biết tháng đó có bao nhiêu ngày
Input - Output
1 = 31
2 = 28
3 = 31
4 = 30
5 = 31
6 = 30
7 = 31
8 = 31
9 = 30
10 = 31
11 = 30
12 = 31
'''
month = int(input("Please input month: "))
if month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month ==12:
    print("31 days")
elif month == 4 or month == 6 or month == 9 or month == 11:
    print("30 days")
elif month == 2:
    print("28 days")
else:
    print("Số không khả dụng, vui lòng chỉ input số từ 1 tới 12")
```
